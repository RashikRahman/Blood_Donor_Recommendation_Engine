```python
df = pd.read_csv('Donor.csv')
```

```python
import random
df['Request']=random.sample(range(30, 140), 109) 
df['Donate']= df['Request'] - random.sample(range(1, 25), 1) 
```

```python
df.head()
```

```python
df['Availability'] = df['Availability'].map({'Y':1,'N':0})
```

```python
df.head()
```

```python
df.describe(include='O')
```

```python
df.to_csv('donor_data.csv', index=False)
```

```python
df = pd.read_csv('donor_data.csv')
df.head()
```

```python
df['rating'] = df['Donate']/df['Request']
```

```python
df = df.sort_values(by='rating',ascending=False)
df
```

```python
blood = 'B-'
top10 = df[df['Blood Group']==blood] 
top10[['Name','Blood Group','Phone']]
```

```python
your_loc = 1207
blood = 'B-'
near_you = df[ (df['Blood Group']==blood) & ((df['Postal Code']<your_loc+8) & (df['Postal Code']>your_loc-8)) ] 
near_you[['Name','Blood Group','Phone']]
```
